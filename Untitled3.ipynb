```python
import cv2
import numpy as np
from ultralytics import YOLO

# Load YOLO model
model = YOLO('/content/ultralytics/runs/detect/train7/weights/best.pt')

# Open video capture
video_path = '/content/video/test2.mp4'
cap = cv2.VideoCapture(video_path)

# Get video properties
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define the codec and create a VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
output_path = '/content/output_video_fill_poly_adjacent_color2.avi'
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Process each frame in the video
while True:
    ret, frame = cap.read()
    if not ret:
        break  # Break the loop if the video is finished

    # Process the frame using YOLO model
    yolov8_results = model(frame)

    for result in yolov8_results:
        boxes = result.boxes.cpu().numpy()

        for xywh in boxes.xywh:
            x, y, w, h = map(int, xywh)

            # Sample color from a single pixel in the adjacent region
            adjacent_color = tuple(frame[y, max(0, x - w // 2)].astype(int))

            # Create vertices of the polygon
            pts = np.array([(max(0, x - w // 2), max(0, y - h // 2)),
                            (min(width, x + w // 2), max(0, y - h // 2)),
                            (min(width, x + w // 2), min(height, y + h // 2)),
                            (max(0, x - w // 2), min(height, y + h // 2))], dtype=np.int32)

            # Reshape the array to match fillPoly's requirements
            pts = pts.reshape((-1, 1, 2))

            # Fill the polygon with the sampled color
            cv2.fillPoly(frame, [pts], tuple(map(int, adjacent_color)))

            # Extract ROI and apply blur
            roi = frame[max(0, y - h // 2):min(height, y + h // 2), max(0, x - w // 2):min(width, x + w // 2)]
            blurred_roi = cv2.GaussianBlur(roi, (15, 15), 0)
            frame[max(0, y - h // 2):min(height, y + h // 2), max(0, x - w // 2):min(width, x + w // 2)] = blurred_roi

    # Write the frame to the output video
    out.write(frame)

# Release resources
cap.release()
out.release()
```
